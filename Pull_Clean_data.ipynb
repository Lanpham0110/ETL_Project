```python
#pulling table/data from Fat_Supply_Quantity_Data.csv

Bdata = pd.read_csv("data.csv")
Bdata
```

```python
Bdata.columns
```

```python
#pulling 2016 data only

Bdata_df = Bdata[['Unnamed: 0', '2016', '2016.1', '2016.2']]
Bdata_df
```

```python
#rename headers

Bdata_df= Bdata_df.rename(columns={'Unnamed: 0': 'Country', '2016':'Obesity (both_sexes)', '2016.1':'Obesity (Male)', '2016.2':'Obesity (Female)' })
Bdata_df
```

```python
#drop row with NaN values

Bdata_df=Bdata_df.dropna()
Bdata_df
```

```python
#drop row with index 2

Bdata_df=Bdata_df.drop([2])
Bdata_df
```

```python
#remove all values inside [ ] and the bracket itself

Bdata_df['Obesity (both_sexes)']= Bdata_df['Obesity (both_sexes)'].str.split("\s+\[").str[0]
Bdata_df['Obesity (Male)']= Bdata_df['Obesity (Male)'].str.split("\s+\[").str[0]
Bdata_df['Obesity (Female)']= Bdata_df['Obesity (Female)'].str.split("\s+\[").str[0]
Bdata_df.head(10)
```

```python
# Reset index 
Bdata_df = Bdata_df.reset_index(drop=True)
Bdata_df
```

```python
#Save table as csv file called: Obesity_percentage

Bdata_df.to_csv('Obsesity_percentage.csv')
```
